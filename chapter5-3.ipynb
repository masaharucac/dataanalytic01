```python
import pandas as pd

# CSVファイルの読み込み
df = pd.read_csv('chap5.csv')

# 平均値を計算
colmean = df.mean()

# 欠損値を平均値で埋める
df2 = df.fillna(colmean)

# 欠損値が残っているか確認
missing_data = df2.isnull().any(axis=0)
print(missing_data)
```

```python
import pandas as pd

# CSVファイルの読み込み
df = pd.read_csv('chap5.csv', encoding='shift-jis')

# 平均値を計算
colmean = df.mean()

# 欠損値を平均値で埋める
df2 = df.fillna(colmean)

# 欠損値が残っているか確認
missing_data = df2.isnull().any(axis=0)
print(missing_data)
```
