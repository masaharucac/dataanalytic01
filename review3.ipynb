```python
import pandas as pd
from sklearn import tree
from sklearn.model_selection import train_test_split
%matplotlib inline
df=pd.read_csv('Survived.csv')
```

```python
df.isnull().any()
```

```python
df.isnull().sum()
```

```python
is_null=df['Age'].isnull()
df.loc[(df['Survived']==0) & (df['Pclass']==1) & (is_null) , 'Age']=40.9
df.loc[(df['Survived']==0) & (df['Pclass']==2) & (is_null) , 'Age']=33.3
df.loc[(df['Survived']==0) & (df['Pclass']==3) & (is_null) , 'Age']=27.4
df.loc[(df['Survived']==1) & (df['Pclass']==1) & (is_null) , 'Age']=34.8
df.loc[(df['Survived']==1) & (df['Pclass']==2) & (is_null) , 'Age']=26.1
df.loc[(df['Survived']==1) & (df['Pclass']==3) & (is_null) , 'Age']=23.2
```

```python
df['Embarked']=df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
x=df[['Pclass','Age','SibSp','Parch','Fare']]
t=df['Survived']
```

```python
def learn (x,t,depth=3):
    x_train,x_test,y_train,y_test=train_test_split(x,t,random_state=0,test_size=0.2)
    model=tree.DecisionTreeClassifier(random_state=0, max_depth=depth,class_weight='balanced')
    model.fit(x_train,y_train)
    score=model.score(x_train,y_train)
    score2=model.score(x_test,y_test)
    return round(score,3),round(score2,3),model
```

```python
df2=pd.DataFrame(columns=['深さ','scoreデータでの正解率','testデータでの正解率'])
for j in range(1,15):
    result=learn(x,t,depth=j)
    ldf=pd.DataFrame({'深さ':j,'scoreデータでの正解率':[result[0]],'testデータでの正解率':[result[1]]})
    df2=pd.concat([df2,ldf],axis=0)
    print(f'深さ・・{j},scoreデータでの正解率・・{result[0]},testデータでの正解率・・{result[1]}')
df2
```
