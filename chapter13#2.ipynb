```python
import pandas as pd
df = pd.read_csv('cinema.csv')
df = df.fillna(df.mean())
x = df.drop(['cinema_id','sales'], axis=1)
t = df['sales']

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,t,random_state=0, test_size=0.3)

from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)

from sklearn.metrics import mean_squared_error
pred = model.predict(x_train)
mse = mean_squared_error(pred, y_train)
mse
```

```python
import math
math.sqrt(mse)
```

```python
from sklearn.metrics import mean_absolute_error
yosoku = [2,3,5,7,11,13]
target = [3,5,8,11,16,19]

mse = mean_squared_error(yosoku,target)

print('rmse:{}'.format(math.sqrt(mse)))
print('mae:{}'.format(mean_absolute_error(yosoku,target)))


yosoku = [2,3,5,7,11,13,46]
target = [3,5,8,11,16,19,23]

mse = mean_squared_error(yosoku,target)

print('mse:{}'.format(math.sqrt(mse)))
print('mae:{}'.format(mean_absolute_error(yosoku,target)))
```

```python
df = pd.read_csv('Survived.csv')
df = df.fillna(df.mean(numeric_only=True))
df.head()
```

```python
x = df[['Pclass', 'Age']]
t = df['Survived']
```

```python
from sklearn import tree

model = tree.DecisionTreeClassifier(random_state=0, class_weight='balanced')
model.fit(x,t)
```

```python
from sklearn.metrics import classification_report
pred = model.predict(x)
out_put = classification_report(y_pred=pred, y_true=t)
print(out_put)
```

```python
out_put2 = classification_report(y_pred=pred, y_true=t, output_dict=True)
pd.DataFrame(out_put2)
```
