```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv('cinema.csv')
df = df.drop('cinema_id', axis=1)
df.head(3)
```

```python
df2 = df.fillna(df.mean())
```

```python
sc_model=StandardScaler()
sc_model.fit(df2)
sc_df = sc_model.transform(df2)
sc_df = pd.DataFrame(sc_df, columns = df.columns)
```

```python
model = PCA(whiten = True)

model.fit (sc_df)
new_df = model.transform (sc_df)
```

```python
new_df = pd.DataFrame(new_df)
col = ['PC1', 'PC2', 'PC3', 'PC4', 'PC5']
new_df.columns = col
new_df
```

```python
ratio = model.explained_variance_ratio_
ratio
```

```python
array = []
for i in range(len(ratio)):
    array.append(sum(ratio[:(i+1)]))
    if array[-1]>=0.85:
        break

i+1
```

```python
tmp_df = pd.concat([sc_df, new_df], axis=1)
tmp_df.head(5)
```

```python
corr_df = tmp_df.corr()
corr_df
```

```python
df_corr = corr_df.loc[:'sales', 'PC1':'PC3']
df_corr
```

```python
for j in df_corr.columns:
    df_corr_sorted = df_corr[j].map(abs).sort_values (ascending =False)
    print(df_corr_sorted)
```
