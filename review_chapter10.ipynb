```python
df_iris=pd.read_csv('iris.csv')
df_iris_no=df_iris.dropna()
```

```python
x=df_iris_no[['がく片幅','花弁長さ','花弁幅']]
t=df_iris_no['がく片長さ']

from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x,t)
```

```python
pred=model.predict(df_iris[df_iris['がく片長さ'].isnull()].loc[:,'がく片幅':'花弁幅'])
df_iris.loc[df_iris['がく片長さ'].isnull(),'がく片長さ']=pred
```

```python
from sklearn.covariance import MinCovDet
df4=df3.loc[:,'atemp':'windspeed']
df4=df4.dropna()
mcd=MinCovDet(random_state=0, support_fraction=0.7)
mcd.fit(df4)

distance=mcd.mahalanobis(df4)
distance
```

```python
distance=pd.Series(distance)
distance
```

```python
distance.plot(kind='box')
```

```python
tmp=distance.describe()
tmp
```

```python
iqr=tmp['75%']-tmp['25%']
kagen=tmp['25%']-(iqr)*1.5
jogen=tmp['75%']+(iqr)*1.5
```

```python
kagen
```

```python
outliner=distance[(distance<kagen) | (distance>jogen)].index
outliner
```

```python
df5=df3.drop(index=outliner)
df5.shape
```
